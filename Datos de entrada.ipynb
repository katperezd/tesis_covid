### CFR
$ CFR= \frac{\text { Number of deaths from disease }}{\text { Number of deaths from disease }+\text { Number of recovered from disease }} \times 100$

```python
suma_muertes = sum(valores_dias_fallecidos)
suma_recuperados = sum(valores_dias_recuperados)
print('Total Fallecidos por Covid-19 a la Fecha:', suma_muertes, '\nTotal Recuperados por Covid-19 a la Fecha:',suma_recuperados)

CFR =  (suma_muertes*100)/(suma_muertes+suma_recuperados)
print('CFR:', CFR)
```

# 3. LOS

```python
from scipy.stats import fisk
# 5 (IQR 39) days
median   = 5
iqr      = 6
max_los  = 200 #maximo tiempo de estadia
# tamao de los pacientes a simular
shape    = 10
location = 1/median
scale    = 28
c        = 3.09
```

```python
#log-logdistic
r = fisk.rvs(c,loc=location,scale=scale,size=shape )
```

```python
distribucion_los = []
for dato in r:
    if(dato>max_los):
        distribucion_los.append(max_los)
    else:
        distribucion_los.append(dato)
distribucion_los = np.array(distribucion_los)
```
