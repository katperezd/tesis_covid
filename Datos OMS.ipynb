```python
def autolabel(rects,bar_label):
    bar_label = np.array(bar_label).astype(str)
    for idx,rect in enumerate(rects):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 1.025*height,bar_label[idx],ha='center', va='bottom', rotation=0)
```
